## Check existing jobs

Each job has a unique job ID. You can check the status of your jobs using the `get_job_status` method. This will return a list of job statuses, which you can iterate through to find the status of each job.


````python
job_statuses = client.get_my_jobs_status()
if job_statuses:
    print("Existing active jobs:")
    for s in job_statuses:
        msg_slices = [
            "",
            f"Job {s.job_id:>6}",
            f"status={s.status:<12}",
            f"service={s.service_name}",
            f"qc_setups={s.qc_setup_list}",
        ]
        print("  ".join(msg_slices))
else:
    print("No active jobs.")
````

````python
# Port Number given when the job is running
job_statuses[0]
````

For each job, a port number is assigned **only when the job is running**.
Then you can use the port number to connect to access the machine to control QPU. 
You can check the port number of a job using the `port_number` attribute of the job status.


````python
if job_statuses[0].port_number:
    print(
        f"Port number for job {job_statuses[0].job_id} is {job_statuses[0].port_number}"
    )
else:
    print(f"Job {job_statuses[0].job_id} is not running, no port number available.")
````
