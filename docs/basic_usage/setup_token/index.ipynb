# Setup Token

## Create a token for authentication

To get a token, You will need to contact the QC-Test Space team to get VPN access to the inner website. After you have VPN access, you can register an account on the QC-Test Space website and obtain a token. The token is used for authentication when using the QC-Test Space Client API.

- Step 1: Go to Dashboard

![Go To Dashboard](./step_01_02.png)

- Step 2: Go to My Account and Create Token

![Go To My Account](./step_03_05.png)

- Step 3: Copy Token

## For Multiple Tokens

We allow users to save multiple tokens for different channels. You can specify the channel name when saving a token. If you don't specify a channel name, the token will be saved under the default channel.

```python
# QCTSSClient.save_token(
#     token="another-token",
#     channel_name="another-channel",
# )
```
